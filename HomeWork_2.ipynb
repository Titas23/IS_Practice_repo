```python

import pandas as pd
```

```python
df = pd.read_csv('Hotel Reservations.csv')
df.head()
```

```python
df.isnull()
```

```python

df.fillna(value= 0, inplace= True)
df.head()
```

```python

from sklearn.preprocessing import LabelEncoder

# changing any string value into numerical value

le = LabelEncoder()

label = le.fit_transform(df['booking_status'])
label_1 = le.fit_transform(df['Booking_ID'])
label_2 = le.fit_transform(df['type_of_meal_plan'])
label_3 = le.fit_transform(df['room_type_reserved'])
label_4 = le.fit_transform(df['market_segment_type'])

df.drop('booking_status', axis=1, inplace=True)
df.drop('Booking_ID', axis=1, inplace=True)
df.drop('type_of_meal_plan', axis=1, inplace=True)
df.drop('room_type_reserved', axis=1, inplace = True)
df.drop('market_segment_type', axis=1, inplace = True)

df['booking_status']= label
df['Booking_ID']= label_1
df['type_of_meal_plan']= label_2
df['room_type_reserved']= label_3
df['market_segment_type']= label_4


df.head(50)

```

```python
from sklearn import preprocessing

df_scaled = preprocessing.scale(df)

df_scaled = pd.DataFrame(df_scaled,columns = df.columns)
```

```python
df_scaled['booking_status'] = df['booking_status']
# removing booking status from the columns
df = df_scaled

df
```

```python
old_cols = df.columns.values
# moving the first four columns around
new_cols =['no_of_adults','no_of_weekend_nights','no_of_children','no_of_week_nights']
df = df.reindex(columns = new_cols)

df
```
